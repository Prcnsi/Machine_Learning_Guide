## *문제 - mean compactness 칼럼 값이 0.07864인 행을 조회해서 인덱스를 얻고 그 인덱스로 그 행의 mean concave points 칼럼 값 조회
-  값에 해당하는 행을 변수에 담고 그 변수로 새 변수에 접근

```python
# 팁
#df[df[칼럼]=값] - 값 조회/검색
#df.loc[i,] - 해당행 출력

# 데이터 조회
# df.iloc[행,열] - 숫자
# df.loc[행,칼럼이름] 

# 데이터 필터링
# df[df[칼럼]==값]
```

```python
# 정답
tmp = cancer_df[cancer_df['mean compactness']==0.07864]
tmp.loc[1, 'mean concave points']

# 참고로 반복문에서 접근시
i = cancer_df[cancer_df['mean compactness']==0.07864]
cancer_df.loc[i, 'mean concave points']
```
