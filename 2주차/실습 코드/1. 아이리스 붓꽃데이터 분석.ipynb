# 1. 붓꽃 데이터 품종 분류(분류)
###  1) 품종은 0,1,2로 이루어져 있음
- sepal: 꽃받침
- petal: 꽃잎

###  2) 칼럼 분석
- sepal length: 꽃받침 길이
- sepal width: 꽃받침 너비
- petal length: 꽃잎 길이
- target: 품종

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

## 1.1 데이터 불러오기

```python
iris=load_iris()
iris
```

```python
# 보기 편한 데이터프레임 형태로 변환
iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
# 타겟값 추가해주기(데이터를 직접 read_csv로 불러오면 보통 타겟값이 있지만 여기서는 내장 데이터라 직접 추가해주는거)
iris_df['label']=iris.target
```
